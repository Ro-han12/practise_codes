```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
import numpy as np
x = np.random.normal(size=1000)

#create normal distribution histogram
sns.displot(x)
```

```python
#create normal distribution curve
sns.displot(x, kind='kde')
```

```python
np.random.seed(0)

#create data
x = np.random.normal(size=1000)

#create normal distribution curve
sns.displot(x, kde=True)
```
