```python
import matplotlib.pyplot as plt 
x=[0,6]
y=[0,250]
plt.plot(x,y)
plt.title("X VS Y")
plt.show()
```

```python
x=[2,4,6,8]
y=[3,9,12,15]
plt.plot(x,y,'o')
```

```python
x=[0,6]
y=[0,250]
plt.plot(x,y,marker='o')
plt.title("X VS Y")
plt.show()
```

```python
import numpy as np
y=np.array([1,4,10])
plt.plot(y,linestyle='dashed')
plt.show()
```

```python
y=np.array([3,8,6,10])
plt.plot(y,marker='*',c='red')
plt.show()
```

```python
x=np.array([1,2,3,5])
y=np.array([10,12,14,16])
plt.subplot(1,2,2)
plt.plot(x,y)
x=np.array([10,20,30,40])
y=np.array([2,4,6,8])
plt.subplot(1,2,1)
plt.plot(x,y)
```

```python
x=np.array([17,23,78,45,66])
y=np.array([99,1,87,33,53])
plt.scatter(x,y)
plt.plot(x,y)
```
